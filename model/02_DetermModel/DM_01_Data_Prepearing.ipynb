```python
# Import potřebných knihoven
import pandas as pd
import numpy as np
```

### Načtení reálných dat a dat digitálního kladečského plánu

```python
# Soubor je načten a přiřazen do proměnné ‚df_real"
other_path = '../../data/raw/ready_timelaps.csv'
df_real = pd.read_csv(other_path, header=0)

# Soubor je načten a přiřazen do proměnné ‚df_sim"
other_path = '../../data/raw/ARSS_plan.csv'
df_sim = pd.read_csv(other_path, header=0)
```

```python
# Zobrazení prvních 5 řádků datasetu
print('Prvních 5 řádků datového rámce')
df_real.head(5)
```

```python
# Zobrazení prvních 5 řádků datasetu
print('Prvních 5 řádků datového rámce')
df_sim.head(5)
```

```python
# Tvorba názvů sloupců
headers = ['id', 'type_brick', 'rotation', 'x', 'y', 'z', 'layer', 'pallet', 'dist']
print('headers\n', headers)
```

```python
# Nahrazení názvů sloupců a následná kontrola datového rámce
df_sim.columns = headers
df_sim.head()
```

```python
# Změna datového typu sloupců
df_sim[['dist','x','y','z']] = df_sim[['dist','x','y','z']].astype(int)
df_sim = df_sim.drop('type_brick', axis=1)
df_sim.head()
```

### Sloučení datových sad podle identifikátoru (ID)

```python
# Sloučení datových sad podle identifikátoru (ID)
df = pd.merge(df_sim, df_real, on='id', how='inner')
df.head()
```

```python
# Seřazení sloupců datového rámce
df = df[['id','type_brick', 'type', 'rotation', 'x', 'y', 'z', 'layer', 'pallet', 'dist', 'start_to_verif', 'verif_to_dest', 'dest_to_end', 'total_time']]
```

### Export datové sady do formátu CSV

```python
df.to_csv('../../data/sim/merged_data.csv', index=False)
```
